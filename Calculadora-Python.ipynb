```python
# Atividade 1

num1_texto = input('Digite o primeiro número:')
num2_texto = input('Digite o segundo número: ')

num1 = float(num1_texto)
num2 = float(num2_texto)

soma = num1 + num2 # para somar os números
subtracao = num1 - num2 # para realizar a subtracao
multiplicacao = num1 * num2 # para realizar a multiplicacao

print(f"O valor total da sua soma é: {soma}")
print(f"O valor total da sua subtração é: {subtracao}")
print(f"O valor total da sua multiplicação é: {multiplicacao}")

if num2 != 0:
  divisao = num1 / num2
  print(f"O valor total da sua divisão é: {divisao}")
else:
  print("Não é possível dividir por zero.")
```

```python
# Laço de repetição que permite realizar várias operações consecutivas
while True:
    print("1. Adição (+)")
    print("2. Subtração (-)")
    print("3. Multiplicação (*)")
    print("4. Divisão (/)")
    print("5. Sair do programa")

    opcao = input("Escolha a opção de (1-5): ")

    # Condicional para verificar se o usuário quer sair antes de pedir os números
    if opcao == "5":
        print("\nCalculadora encerrada com sucesso. Até logo!")
        break  # Quebra o laço de repetição e finaliza o programa

    # Condicional que valida se a opção digitada é válida (de 1 a 4)
    if opcao in ["1", "2", "3", "4"]:
        # Utilizando o comando 'input' e convertendo os valores para float (ponto flutuante)
        num1 = float(input("\nDigite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))

        # Estrutura condicional para decidir qual operação matemática realizar
        if opcao == "1":
            resultado = num1 + num2
            print(f"\n[RESULTADO] A soma de {num1} + {num2} é igual a: {resultado}")

        elif opcao == "2":
            resultado = num1 - num2
            print(f"\n[RESULTADO] A subtração de {num1} - {num2} é igual a: {resultado}")

        elif opcao == "3":
            resultado = num1 * num2
            print(f"\n[RESULTADO] A multiplicação de {num1} x {num2} é igual a: {resultado}")

        elif opcao == "4":
            # Validação extra importante: evitar divisão por zero
            if num2 != 0:
                resultado = num1 / num2
                print(f"\n[RESULTADO] A divisão de {num1} / {num2} é igual a: {resultado}")
            else:
                print("\n[ERRO] Não é possível dividir um número por zero!")
    else:
        print("\n[ERRO] Opção inválida! Por favor, escolha um número de 1 a 5.")
```
